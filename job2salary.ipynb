```python
import pandas as pd
import numpy as np

# 数据加载
data = pd.read_excel('./jobs_4k.xls')
data
```

```python
positionName = data['positionName'].tolist()
print(len(positionName))
temp = set(positionName)
print(len(temp))
```

```python
skillLables = data['skillLables'].tolist()
print(len(skillLables))
```

```python
# 图可视化
from collections import defaultdict
skill_position_graph = defaultdict(list)
for p, s in zip(positionName, skillLables):
    skill_position_graph[p] += eval(s)
print(skill_position_graph)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import warnings
warnings.filterwarnings('ignore')
G = nx.Graph(skill_position_graph)
# 30 个随机选取的工作岗位，进行图可视化
sample_nodes = random.sample(positionName, k=30)
sample_nodes_connnections = sample_nodes
for p, skills in skill_position_graph.items():
    if p in sample_nodes:
        sample_nodes_connnections += skills
# 抽取G的节点作为子图
f = plt.figure(figsize=(30,15))
sample_graph = G.subgraph(sample_nodes_connnections)
pos = nx.spring_layout(sample_graph, k=1)
nx.draw(sample_graph, pos, with_labels=True, node_size=30, font_size=10)
plt.show()
f.savefig('job-connections.png',dpi=64)
```

```python
# 对核心能力和职位进行排序（按照影响力）-->PageRank
pr = nx.pagerank(G, alpha=0.9)
ranked_position_and_ability = sorted([(name, value) for name, value in pr.items()], key=lambda x: x[1], reverse=True)
ranked_position_and_ability
```

```python
# 取出特征X, 不包括salary
X_data = data.drop(['salary'], axis=1)
y = data['salary'].tolist()
y
```

```python
# 将特征内容拼接成字符串
X_data['merged'] = X_data.apply(lambda x: ''.join(str(x)), axis=1)
# 转化为list
X_string = X_data['merged'].tolist()
print(X_string[0])
```

```python
import jieba

# 合并
def get_one_row_job_string(x_string_row):
    job_string = ''
    for i ,element in enumerate(x_string_row.split('\n')):
        if len(element.split()) == 2:
            _, value = element.split()
            # i=0 是id字段，舍弃
            if i == 0:
                continue
            # 只保留value
            job_string += value
    return job_string
        
```

```python
import re
def token(string):
    return re.findall('\w+', string)
cut_X = []
for i, row in enumerate(X_string):
    job_string = get_one_row_job_string(row)
    cut_X.append(' '.join(list(jieba.cut(''.join(token(job_string))))))
    #print(cut_X)
```

```python
# 使用TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
# 原始数据的TF-IDF特征
X = vectorizer.fit_transform(cut_X)
print(X[0])
```

```python
y[:5]
```

```python
# 求平均值， 比如薪资 10-25k 
y_numical = [np.mean(list(map(float, re.findall('\d+', s)))) for s in y]
y_numical = np.array(y_numical).reshape(-1,1)
y_numical[:5]
```

```python
# 使用KNN回归
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=2)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y_numical, test_size=0.3)
```

```python
# 模型训练和测试
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error
print('MAE: ', mean_absolute_error(y_pred, y_test))
print('MSE: ', mean_squared_error(y_pred, y_test))
```

```python
def predict_by_label(test_string, model):
    test_words = list(jieba.cut(test_string))
    test_vec = vectorizer.transform(test_words)
    pred_y = model.predict(test_vec)
    return pred_y[0]
```

```python
test = '测试 北京 3年 专科'
test2 = '测试 北京 4年 专科'
test3 = '算法 北京 4年 本科'
test4 = 'UI 北京 4年 本科'
print(test, predict_by_label(test, model))
print(test2, predict_by_label(test2, model))
print(test3, predict_by_label(test3, model))
print(test4, predict_by_label(test4, model))
```
